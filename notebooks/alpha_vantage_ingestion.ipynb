# Alpha Vantage Ingestion

```python
# Imports
import sys
import os
from dotenv import load_dotenv
sys.path.append('..')  # Add parent directory to path

from alpha_vantage.timeseries import TimeSeries
from datetime import datetime, timedelta
from decimal import Decimal
import pandas as pd
from src.models import OHLCVRecord, DataSource

load_dotenv()
```

```python
symbol = "AAPL"
ts = TimeSeries(key=os.getenv('ALPHA_VANTAGE_API_KEY'), output_format='pandas')
data: pd.DataFrame
data, meta_data = ts.get_daily(symbol=symbol, outputsize='compct')

print(data)

```

```python
print(data.head())

data = data.rename(columns={
    'open':'Open',
    'high':'High',
    'low':'Low',
    'close': 'Close',
    'volume': 'Volume'
})

print(data.head())
```

```python
ohlcv_list: list[OHLCVRecord] = []

for index, row in data.iterrows():
    record = OHLCVRecord(
        timestamp = index.to_pydatetime(),
        symbol="AAPL",
        source=DataSource.YAHOO_FINANCE,
        open=row["Open"],
        high=row["High"],
        low=row["Low"],
        close=row["Close"],
        volume=int(row["Volume"]),
        adjusted_close=None,
        metadata={}
    )
    ohlcv_list.append(record)


```
